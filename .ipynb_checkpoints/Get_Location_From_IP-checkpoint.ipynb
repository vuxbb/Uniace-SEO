```python
def retrieve_city_from_ip(ip):
    city = []
    for i in range(len(ip)):
        reader = geoip2.database.Reader('C:\\Data\\mnt\\nhdd\\2022\\checkip\\GeoIP2-City_20220308\\GeoIP2-City.mmdb')
        try :
            data = reader.city(ip[i])
            #print(data.city.names['en'])
            city.append(data.city.names['en'])
        except:
            #print('The address {} is not in the database'.format(ip[i]))
            city.append('The address is not in the database'.format(ip[i]))
    data = {'IP':ip,'City':city}
    return data
```
